```python
class ReturnValue:
    def __init__(self, clusters_pos_, labels_):
        self.clusters_pos_ = clusters_pos_
        self.labels_ = labels_
    
        
class Kmean:
    """choose the number of clusters"""
    def __init__(self,k):
        self.k = k  #nbre de groupes(cluster)    
    def fit(self,X):
        """randomly choose k centroids"""
        c=X[np.random.choice(len(X),self.k,replace=False),:]
        """find index cluster"""
        lab=np.apply_along_axis(lambda x: np.argmin(np.apply_along_axis(np.linalg.norm, 1, x-c)),1,X)
        
        t=True
        while t:
            """calculate new centroids"""
            l=[]
            for i in range(self.k):
                l.append(np.mean(X[lab==i], axis=0).tolist())
            cn=np.array(l)    
            """find index cluster for the new centroids"""
            labn=np.apply_along_axis(lambda x: np.argmin(np.apply_along_axis(np.linalg.norm, 1, x-cn)),1,X)
            if sum(labn==lab)==len(labn):
                t=False
            else:
                lab=labn
        #return {'clusters_pos_': cn, 'labels_': labn}
        return ReturnValue(cn,labn)
     
```

**fit** : take data X  as input and generates attributes **labels_** and **clusters_pos_**

```python
km=Kmean(3)
kmX=km.fit(X)
```

```python
labn=kmX.labels_
cn=kmX.clusters_pos_
```

```python
def plot_kmean(k,labn,cn):
    for i in range(k):
        sns.scatterplot(x=X[labn==i][:,0],y=X[labn==i][:,1])
       
    sns.scatterplot(x=cn[:,0],y=cn[:,1],s=60)
    
plot_kmean(3,labn,cn)    
```

**Note:**   
* k-means is guaranteed to converge but the final cluster configuration to which it converges is not in general unique, and depends on the initial centroid locations.

**how do you initialize the centroids in k-means?**

One straightforward way of doing it is to pick them **randomly** from among your data points. This is does not work particularly well, however, because it becomes overwhelmingly likely that many of initial centroids will end up in the same true cluster.
A better idea, implemented in the **farthest heuristic**, is to initialize the first centroid randomly, but then to initialize the second centroid to be the data point farthest away from it.
